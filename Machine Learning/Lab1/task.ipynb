```python
# This practical task has been written by Ayat Ospanov, group #317, CMC, MSU. 2014.
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot
import sklearn as skl
import sklearn.neighbors
import sklearn.grid_search
import sklearn.cross_validation
import sklearn.metrics
%matplotlib inline
```

```python
def genData(numOfElemInClass, numOfClasses):
    data = np.array([np.array([0, 0]), 0])
    for classNum in range(numOfClasses):
        mju = np.array([0, 3])
        sigma = np.array([1, 2])
        for i in range(numOfElemInClass):
            x, y = np.random.normal(mju[classNum], sigma[classNum]), np.random.normal(mju[classNum], sigma[classNum])
            data = np.vstack((data, np.array([np.array([x, y]), classNum])))
    return np.delete(data, 0, 0)
```

```python
def plotData(data):
    mpl.pyplot.scatter([data[i][0][0] for i in range(len(data))],
               [data[i][0][1] for i in range(len(data))],
               c = [data[i][1] for i in range(len(data))],
               s = 50,
               cmap = mpl.colors.ListedColormap(['#FF0000', '#00FF00']))
    fig = mpl.pyplot.gcf()
    fig.set_size_inches(fig.get_size_inches() * 2)
    mpl.pyplot.show()
```

```python
def countMesh(data):
    data_x = [data[i][0][0] for i in range(len(data))]
    data_y = [data[i][0][1] for i in range(len(data))]
    x_max = np.max(data_x)
    x_min = np.min(data_x)
    y_max = np.max(data_y)
    y_min = np.min(data_y)
    return np.meshgrid(np.arange(x_min - 1, x_max + 1, 0.08),
                       np.arange(y_min - 1, y_max + 1, 0.08))
```

```python
def countMeshLabels(mesh_x, mesh_y):
    return knn.predict([[mesh_x.ravel()[i], mesh_y.ravel()[i]]
                        for i in range(len(mesh_x.ravel()))])
```

```python
def plotDataOnMesh(data):
    data_x = [data[i][0][0] for i in range(len(data))]
    data_y = [data[i][0][1] for i in range(len(data))]
    mesh_x, mesh_y = countMesh(data)
    mesh_labels = countMeshLabels(mesh_x, mesh_y)
    mpl.pyplot.pcolormesh(mesh_x, mesh_y, mesh_labels.reshape(mesh_x.shape),
                          cmap = mpl.colors.ListedColormap(['#FF00FF', '#00FFFF']))
    mpl.pyplot.scatter(data_x, data_y, c = [data[i][1] for i in range(len(data))],
                       s = 50, cmap = mpl.colors.ListedColormap(['#FF0000', '#00FF00']))
    fig = mpl.pyplot.gcf()
    fig.set_size_inches(fig.get_size_inches() * 2)
    mpl.pyplot.show()
```

```python
data = genData(100, 2)
```

```python
plotData(data)
```

## Выбор наилучшего параметра с помощью 5-fold cross-validation

```python
knnForGS = skl.neighbors.KNeighborsClassifier()
parameters = {'n_neighbors' : np.arange(1, 51)}
clf = skl.grid_search.GridSearchCV(knnForGS, parameters, cv = 5)
clf.fit([[data[i][0][0], data[i][0][1]] for i in range(len(data))],
        [data[i][1] for i in range(len(data))])
```

```python
print('Лучший параметр k для данной выборки: ', end='')
print(clf.best_params_['n_neighbors'], end='\n\n')
```

```python
def plotCV_k(data):
    fig = mpl.pyplot.gcf()
    fig.set_size_inches(fig.get_size_inches() * 2)
    mpl.pyplot.plot([data[i][0]['n_neighbors'] for i in range(len(data))], [data[i][1] for i in range(len(data))])
```

```python
def plotCV_k_intervals(data):
    fig = mpl.pyplot.gcf()
    fig.set_size_inches(fig.get_size_inches() * 2)
    mpl.pyplot.plot([data[i][0]['n_neighbors'] for i in range(len(data))], [data[i][1] for i in range(len(data))])
    for i in range(len(data)-1):
        mpl.pyplot.fill_between([data[i][0]['n_neighbors'], data[i+1][0]['n_neighbors']],
                               [data[i][1] - data[i][2].std(), data[i+1][1] - data[i+1][2].std()],
                               [data[i][1] + data[i][2].std(), data[i+1][1] + data[i+1][2].std()],
                               alpha = 0.3, facecolor='green')
```

```python
plotCV_k_intervals(clf.grid_scores_)
```

## Вычисление качества на контроле метода 5NN

```python
train_data, test_data, train_y, test_y = \
skl.cross_validation.train_test_split([[data[i][0][0], data[i][0][1]] for i in range(len(data))],
                                      [data[i][1] for i in range(len(data))],
                                      test_size=0.2)
```

```python
knn_test = skl.neighbors.KNeighborsClassifier(5)
knn_test.fit(train_data, train_y)
acc_before = skl.metrics.accuracy_score(test_y, knn_test.predict(test_data))
acc_before
```

#### Добавление веса ко второму признаку

```python
mul_train_data = np.copy(train_data)
mul_test_data = np.copy(test_data)
for i in range(len(mul_train_data)):
    mul_train_data[i][1] *= 100
for i in range(len(mul_test_data)):
    mul_test_data[i][1] *= 100
```

```python
knn_test.fit(mul_train_data, train_y)
acc_after = skl.metrics.accuracy_score(test_y, knn_test.predict(mul_test_data))
acc_after
```

```python
print("Качество метода снизилось на ", end = '')
print((acc_before - acc_after) * 100, end = "%. ")
print("Это связано с тем, что метод k ближайших соседей чувствителен к масштабу признаков. Это означает, что перед тем, как использовать метод kNN, нужно нормировать признаки.")
```
