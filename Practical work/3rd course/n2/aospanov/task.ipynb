```python
def euclidMetric(x, y):
    x = x.astype(np.int32)
    y = y.astype(np.int32)
    yn, xn = np.meshgrid((y ** 2).sum(axis=1), (x ** 2).sum(axis=1))
    return np.sqrt(xn + yn - 2 * x.dot(y.T))
```

```python
def features(maxIndx, featNum):
    return np.random.choice(maxIndx, featNum, replace=False)
```

```python
def kNN(train, test, trainLabels, neighborsNum = 5, blockSize = 500, name = "brute", met="minkowski"):
    if name is "myKNN":
        tmp = np.empty((train.shape[0], test.shape[0]))
        for i in range(0, test.shape[0], blockSize):
            indxL = i
            indxR = indxL + blockSize
            tmp[:, indxL : indxR] = euclidMetric(train, test[indxL : indxR])
        tmp.sort(axis=0)
        return tmp[: neighborsNum, :]
    elif name is "brute" or name is "kd_tree":
        nbrs = sklearn.neighbors.NearestNeighbors(n_neighbors = neighborsNum,
                                                  algorithm = name, metric = met).fit(train)
        tmp = np.empty((test.shape[0], neighborsNum), dtype = np.int32)
        for i in range(0, test.shape[0], blockSize):
            indxL = i
            indxR = indxL + blockSize
            tmp[indxL : indxR, :] = nbrs.kneighbors(test[indxL : indxR])[1]
        return trainLabels[tmp]
    else:
        return "Error"
```

## 5th task

```python
def KFold(n, n_fold, random = 0):
    if random:
        arr = np.random.choice(n, n, replace=False)
    else:
        arr = np.arange(n)

    if n % n_fold is not 0:
        last = n // n_fold + 1
    else:
        last = n // n_fold
    tup = (arr[last :], arr[: last])
    res = [tup]
    k = last
    for i in range(1, n_fold):
        n -= k
        n_fold -= 1
        if n % n_fold is not 0:
            k = n // n_fold + 1
        else:
            k = n // n_fold
        res.append((np.append(arr[: last], arr[last + k :]),
                    arr[last : last + k]))
        last += k
    return res
```

## 6th task

```python
def runLengthEncoding(x):
    x = np.sort(x)
    d = np.diff(x)
    a = (np.array(np.where(d != 0)) + 1).flatten()
    if (a.size != 0):
        b = np.insert(np.diff(a), 0, a[0])
        return np.append(x[np.where(d != 0)], x[x.size - 1]), \
            np.append(b, d.size - a[a.size - 1] + 1)
    else:
        return np.array([x[0]]), np.array([x.size])
```

```python
def kNNScore(k, neighborsLbls, testLabels):
    lbls = np.empty(neighborsLbls.shape[0], dtype=np.int32)
    for i in range(neighborsLbls.shape[0]):
        tmp = runLengthEncoding(neighborsLbls[i][:k])
        lbls[i] = tmp[0][np.argmax(tmp[1])]
    return np.sum(lbls == testLabels) / neighborsLbls.shape[0]
```

```python
def scoreCV(k, CV, cvIndex, data, trainLbls, features, metr = "minkowski"):
    scoresForK = np.empty([k.size, CV])
    i = 0;
    for (train, test) in cvIndx:
        stime = time.clock()
        neighIndx = kNN(data[train][:, features], data[test][:, features],\
                        trainLbls[train], k[k.size - 1], 3500,\
                        name = "brute", met = metr)
        print("kNN with", metr, "metric (Iter.", i, "):", time.clock() - stime)
        for curK in range(k.size):
            scoresForK[curK][i] = kNNScore(k[curK], neighIndx, trainLbls[test])
        i += 1
    avg = np.average(scoresForK, axis = 1)
    max = np.max(scoresForK, axis = 1)
    return (avg, max - avg)
```

## 8th task

```python
def scoreCVWeighted(k, CV, cvIndex, data, trainLbls, features, eps = 0.01, metr = "minkowski"):
    scoresForK = np.empty([k.size, CV])
    it = 0;
    neighborsNum = k[k.size - 1]
    blockSize = 3500
    for (train, test) in cvIndx:
        stime = time.clock()
        nbrs = sklearn.neighbors.NearestNeighbors(n_neighbors = neighborsNum, algorithm = 'brute',
                                                  metric = metr).fit(data[train][:, features])
        tmp1 = np.empty((test.shape[0], neighborsNum), dtype = np.int32)
        tmp2 = np.empty((test.shape[0], neighborsNum), dtype = np.int32)
        for i in range(0, test.shape[0], blockSize):
            indxL = i
            indxR = indxL + blockSize
            tmp = nbrs.kneighbors(data[test][indxL : indxR, features])
            tmp1[indxL : indxR, :] = tmp[0]
            tmp2[indxL : indxR, :] = tmp[1]

        labels = trainLbls[train][tmp2]
        dist = tmp1

        print("Weighted kNN with", metr, "metric (Iter.", it, "):", time.clock() - stime)

        for curK in range(k.size):
            dIndx = np.zeros(labels.shape[0])
            for i in range(labels.shape[0]):
                sum = np.zeros(10) # max number of class
                for j in range(k[curK]):
                    sum[labels[i][j]] += 1 / (dist[i][j] + eps)
                dIndx[i] = np.argmax(sum)
            scoresForK[curK][it] = np.sum(dIndx == trainLbls[test]) / dIndx.size
        it += 1
    avg = np.average(scoresForK, axis = 1)
    max = np.max(scoresForK, axis = 1)
    return (avg, max - avg)
```

```python
def kNN(train, test, trainLabels, neighborsNum = 10, blockSize = 3500, name = "brute", met="minkowski"):
    nbrs = sklearn.neighbors.NearestNeighbors(n_neighbors = neighborsNum,
                                              algorithm = name, metric = met).fit(train)
    tmp = np.empty((test.shape[0], neighborsNum), dtype = np.int32)
    for i in range(0, test.shape[0], blockSize):
        indxL = i
        indxR = min(indxL + blockSize, test.shape[0])
        tmp[indxL : indxR, :] = nbrs.kneighbors(test[indxL : indxR])[1]
    return trainLabels[tmp]
```

```python
def scoreCV(k, CV, cvIndex, data, trainLbls, metr = "minkowski"):
    scoresForK = np.empty([k.size, CV])
    i = 0;
    for (train, test) in cvIndx:
        stime = time.clock()
        neighIndx = kNN(data[train], data[test],\
                        trainLbls[train], k[k.size - 1], 3500,\
                        name = "brute", met = metr)
        print("kNN with", metr, "metric (Iter.", i, "):", time.clock() - stime)
        for curK in range(k.size):
            scoresForK[curK][i] = kNNScore(k[curK], neighIndx, trainLbls[test])
        i += 1
    avg = np.average(scoresForK, axis = 1)
    max = np.max(scoresForK, axis = 1)
    return (avg, max - avg)
```

```python
def scoreCVWeighted(k, CV, cvIndex, data, trainLbls, eps = 0.01, metr = "minkowski"):
    scoresForK = np.empty([k.size, CV])
    it = 0;
    neighborsNum = k[k.size - 1]
    blockSize = 3500
    for (train, test) in cvIndx:
        stime = time.clock()
        nbrs = sklearn.neighbors.NearestNeighbors(n_neighbors = neighborsNum,
                                                  algorithm = 'brute', metric = metr).fit(data[train])
        tmp1 = np.empty((test.shape[0], neighborsNum), dtype = np.int32)
        tmp2 = np.empty((test.shape[0], neighborsNum), dtype = np.int32)
        for i in range(0, test.shape[0], blockSize):
            indxL = i
            indxR = min(indxL + blockSize, test.shape[0])
            tmp = nbrs.kneighbors(data[test][indxL : indxR])
            tmp1[indxL : indxR, :] = tmp[0]
            tmp2[indxL : indxR, :] = tmp[1]

        labels = trainLbls[train][tmp2]
        dist = tmp1

        print("Weighted kNN with", metr, "metric (Iter.", it, "):", time.clock() - stime)

        for curK in range(k.size):
            dIndx = np.zeros(labels.shape[0])
            for i in range(labels.shape[0]):
                summ = np.zeros(trainLbls.size) # max number of class
                for j in range(k[curK]):
                    summ[labels[i][j]] += 1 / (dist[i][j] + eps)
                dIndx[i] = np.argmax(summ)
            scoresForK[curK][it] = np.sum(dIndx == trainLbls[test]) / dIndx.size
        it += 1
    avg = np.average(scoresForK, axis = 1)
    max = np.max(scoresForK, axis = 1)
    return (avg, max - avg)
```

Во всех случаях, и в MNIST, и в NEWSGroups, лучше всего для классификации подошел метод с косинусной метрикой (прирост от 2% до 8%). А взвешенный метод не дал особых выигрышей, но дал (~1-2%), т.о. лучше использовать взвешенный kNN. Количество соседей тоже очень влияло на точность: для MNIST самый лучший результат был для k = 7, а для NEWS лучший результат был при k = 1. В случае NEWS лучше использовать TF-IDF преобразование в паре с косинусной метрикой (дает прирост в 30%)
