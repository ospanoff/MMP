```python
def plot_errors(FPMask, FNMask):
    plt.rcParams['figure.figsize'] = (15, 5)
    FP = np.sum(np.sum(FPMask, axis=-1), axis=-1)
    FN = np.sum(np.sum(FNMask, axis=-1), axis=-1)
    plt.plot(FP, label='FP')
    plt.plot(FN, label='FN')
    plt.xlabel('Frame number', size=13)
    plt.ylabel('Quantity', size=13)
    plt.legend(loc='best')
    plt.show()
```

### Одномерная гауссиана

```python
def plot_errors_threshold(mju, threshold):
    for thold in threshold:
        mask = np.zeros((img.shape[0], img.shape[1], img.shape[2]), dtype=np.bool)
        for i in range(mask.shape[0]):
            mask[i] = (np.abs(gray_img[i] - mju) > thold)

        tmp = mask - target
        FPMask = tmp > 0.999
        FNMask = tmp < -0.999
        del(tmp, mask)
        plt.rcParams['figure.figsize'] = (15, 5)
        plt.title('Threshold (k*sigma) = ' + str(thold), size=13)
        plot_errors(FPMask, FNMask)
        del(FPMask, FNMask)
```

```python
def plot_ROC_1D(mju, threshold):
    FPR = [0]
    TPR = [0]
    P = np.sum(target == 1)
    N = np.sum(target == 0)
    for th in threshold:
        mask = np.zeros((img.shape[0], img.shape[1], img.shape[2]), dtype=np.bool)
        for i in range(mask.shape[0]):
            mask[i] = (np.abs(gray_img[i] - mju) > th)
        
        tmp = mask - target
        FPMask = tmp > 0.999
        FNMask = tmp < -0.999
        FP = np.sum(FPMask)
        FN = np.sum(FNMask)
        TP = P - FN
        TPR.append(TP / P)
        FPR.append(FP / N)
        del(mask, tmp, FPMask, FNMask)
    plt.rcParams['figure.figsize'] = (8, 8)
    plt.ylim(0, 1.1)
    plt.xlim(-0.1, 1.1)
    FPR.append(1)
    TPR.append(1)
    FPR = np.array(FPR)
    TPR = np.array(TPR)
    indx = FPR.argsort()
    plt.plot(FPR[indx], TPR[indx])
    plt.xlabel('FPR', size=12)
    plt.ylabel('TPR', size=12)
    plt.title('AUC = ' + str(np.trapz(TPR[indx], FPR[indx])))
    del(FPR, TPR, indx)
```

### Смесь гауссиан

```python
def genData(numOfClasses, numOfElemInClass, D, coef_=10):
    mean = [i + np.random.random(D) * coef_ for i in range(numOfClasses)]
    cov = np.array(
        [np.diag(np.random.random(D)) for i in range(numOfClasses)])

    X = np.empty((numOfClasses * numOfElemInClass, D))
    for i in range(numOfClasses):
        X[i * numOfElemInClass : (i + 1) * numOfElemInClass, :] =\
            np.random.multivariate_normal(mean[i], cov[i], numOfElemInClass)

    y = np.ones((numOfClasses * numOfElemInClass, 1), dtype=np.int32)
    for i in range(numOfClasses):
        y[i * numOfElemInClass: (i + 1) * numOfElemInClass] = i
    return X, y

def plot_Xy(X, y):
    plt.rcParams['figure.figsize'] = (8, 8)
    plt.scatter(X[:, 0], X[:, 1], s=40, c=y.astype(np.int32), cmap=matplotlib.colors.ListedColormap(
        ['#ff00ff', '#9CFB00', '#00ffff']))
    plt.show()
```

```python
def kMeans(X, K, eps=1e-3):
    y_pre = np.zeros((X.shape[0], 1))
    for i in range(K):
        y_pre[i * (y_pre.shape[0]//K) : (i + 1) * (y_pre.shape[0]//K)] = i
    np.random.shuffle(y_pre)

    tmp = np.zeros((K, X.shape[1]))
    mju = np.zeros((K, X.shape[1]))
    while True:
        for i in range(K):
            mju[i] = np.sum((y_pre == i) * X, axis=0) / np.sum(y_pre == i)
        y_pre = np.argmin([np.linalg.norm(X - mju[i], axis=1) for i in range(K)], axis=0)[:, np.newaxis]
        if np.all(np.abs(tmp - mju)) < eps:
            break
        mju = tmp

    return y_pre, mju
```

##### EM алгоритм на выборке 'Traffic'

```python
def EM(X, K, mju=None, delta=1e-5):
    if mju is None:
        mju = X[np.random.choice(X.shape[0], K, replace=False)]
    D = X.shape[1]
    w = np.ones(K) / K
    S = np.zeros((K, D, D))
    for x in S:
        np.fill_diagonal(x, np.ones(D))
    alpha = np.zeros((X.shape[0], K))
    dens = np.zeros(K)
    steps = 100
    step = 0
    MLE = []
    while True:
        step += 1
        alpha_0 = np.copy(alpha)
        mle = 0

        for i in range(X.shape[0]):
            for k in range(K):
                dens[k] = w[k] * np.exp(-0.5 * np.dot(np.dot(X[i] - mju[k], np.linalg.inv(S[k])), X[i] - mju[k])) /\
                          np.sqrt(((2 * np.pi) ** D) * np.linalg.det(S[k]))

            mle_pre = np.sum(dens)
            alpha[i] = dens / mle_pre
            mle += np.log(mle_pre)
        w = np.mean(alpha, axis=0)
        mju = np.dot(alpha.T, X) / np.sum(alpha, axis=0)[:, np.newaxis]

        for k in range(K):
            S[k] = np.dot((X - mju[k]).T, alpha[:, k, np.newaxis] * (X - mju[k])) / np.sum(alpha[:, k])
            S[k] += np.eye(D) / 10

        if len(MLE) > 0 and np.abs(mle - MLE[len(MLE) - 1]) < delta:
            status = 'good mle'
            break
        MLE.append(mle)
        if (np.max(np.abs(alpha - alpha_0)) < delta):
            status = 'good delta'
            break
        if step > steps:
            status = 'steps limit'
            break

    return {
        'decision': alpha,
        'steps': step,
        'MLE': MLE,
        'w': w,
        'status': status
    }
```
